```python
import shutil
storedir = "/workspace/algorithm/lijia/liyu/DNAMeth/swavDNA/check_100kbratio_linear_lr0.1_nopretrianed/Fold0/cam/test_PredR/"
for i,ifile in enumerate(reffilenames):
    filename = "allc_"+ifile.split(".")[0]+".png"
    if not os.path.exists(os.path.join(storedir,marjortype[i],filename)):
        continue
    if not os.path.exists(os.path.join(storedir,"subtype")):
        os.makedirs(os.path.join(storedir,"subtype"))
    if not os.path.exists(os.path.join(storedir,"subtype",celltype[i])):
        os.makedirs(os.path.join(storedir,"subtype",celltype[i]))
    shutil.copy(os.path.join(storedir,marjortype[i],filename),os.path.join(storedir,"subtype",celltype[i]))
```
